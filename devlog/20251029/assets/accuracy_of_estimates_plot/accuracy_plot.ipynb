```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("results.csv")
```

```python
plt.plot([10**t for t in range(-5,0)], [10**t for t in range(-5,0)], linestyle="dashed", color="grey")
plt.scatter(df["rate"], df["estimated"], color=df["discretized"], zorder=2)
plt.xlabel("True Rate")
plt.ylabel("Estimated Rate")
plt.xscale("log")
plt.yscale("log")
plt.axis("square")
plt.savefig("accuracy.svg")
plt.show()
```
